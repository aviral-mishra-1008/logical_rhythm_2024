```python
from sklearn.model_selection import cross_val_score, GridSearchCV, RandomizedSearchCV
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn import tree
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import (RandomForestClassifier, GradientBoostingClassifier,
                               AdaBoostClassifier, VotingClassifier)
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score
from itertools import combinations
```

```python
def evaluate_model(model, X, y, cv_folds=5):
    cv_scores = cross_val_score(model, X, y, cv=cv_folds, scoring='roc_auc')
    mean_score = cv_scores.mean()
    print(f"{model.__class__.__name__} CV Scores: {cv_scores}")
    print(f"Mean CV Score: {mean_score}")
    return mean_score
```

```python
def tune_model(model, param_grid, X, y, model_name):
    clf = GridSearchCV(model, param_grid, cv=5, verbose=True, n_jobs=-1)
    best_model = clf.fit(X, y)
    print(f"{model_name} Best Score: {best_model.best_score_}")
    print(f"{model_name} Best Parameters: {best_model.best_params_}")
    return best_model
```

```python
def stack_models(models, X, y):
    voting_clf = VotingClassifier(estimators=[(name, model) for name, model in models.items()], voting='soft')
    return evaluate_model(voting_clf, X, y)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import VotingClassifier, StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from sklearn.naive_bayes import GaussianNB
import pandas as pd

classifiers = {
    'GaussianNB': GaussianNB(),
    'LogisticRegression': LogisticRegression(max_iter=200),
    'RandomForest': RandomForestClassifier(random_state=1, n_estimators=10),
    'KNeighborsClassifier': KNeighborsClassifier(),
    'XGBClassifier': XGBClassifier(random_state=1),
    'GradientBoosting': GradientBoostingClassifier(random_state=1),
    'AdaBoost': AdaBoostClassifier(random_state=1),
}

base_classifiers = [
    ('GaussianNB', GaussianNB()),
    ('LogisticRegression', LogisticRegression(max_iter=200)),
    ('RandomForest', RandomForestClassifier(random_state=1, n_estimators=10)),
    ('KNeighborsClassifier', KNeighborsClassifier()),
    ('XGBClassifier', XGBClassifier(random_state=1)),
    ('GradientBoosting', GradientBoostingClassifier(random_state=1)),
    ('AdaBoost', AdaBoostClassifier(random_state=1)),
]
meta_classifier = LogisticRegression(max_iter=200)
voting_clf = VotingClassifier(estimators=base_classifiers, voting='soft')
stacking_clf = StackingClassifier(estimators=base_classifiers, final_estimator=meta_classifier)
for model_name, model in {'Voting': voting_clf, 'Stacking': stacking_clf}.items():    
    pipeline = Pipeline([
        ('scaler', StandardScaler()),
        ('clf', model)
    ])
    
    pipeline.fit(X_train, y_train)
    
    predictions = pipeline.predict(numeric_dftest)
    
    result_df = pd.DataFrame({
        'FILENAME': dftest['FILENAME'],
        'label': predictions
    })
    
    csv_filename = f'pred_{model_name}.csv'
    result_df.to_csv(csv_filename, index=False)
    print(f"Predictions saved to {csv_filename}")
```
